```python
import matplotlib.pyplot as plt

country_counts = df_cleaned['Country'].value_counts()

plt.figure(figsize=(15, 4))
plt.bar(country_counts.index, country_counts.values, color=['skyblue', 'salmon'])
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('CO2_Emissions_tons_per_capita')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
plt.savefig("CO2 distribution.jpg")
```

```python
df_cleaned = df_cleaned.dropna(subset=['CO2_Emissions_tons_per_capita', 'Avg_Temperature_degC'])

plt.figure(figsize=(15, 8))
sns.scatterplot(
    data=df_cleaned,
    x='CO2_Emissions_tons_per_capita',
    y='Avg_Temperature_degC',
    hue='Country',
    palette='tab10'
)

plt.title("COâ‚‚ Emissions vs Average Temperature by Country")
plt.xlabel("COâ‚‚ Emissions (tons per capita)")
plt.ylabel("Avg Temperature (Â°C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig("CO2 vs Avg Temp.jpg")
```

```python
df = pd.read_csv("temp_cleaned.csv")
sns.set(style="whitegrid")
grouped = df.groupby("Country")[["CO2_Emissions_tons_per_capita", "Renewable_Energy_pct"]].mean()
grouped = grouped.sort_values("CO2_Emissions_tons_per_capita", ascending=False)
fig, ax1 = plt.subplots(figsize=(14, 6))
grouped.plot(kind="bar", ax=ax1)
plt.title("Average CO2 Emissions and Renewable Energy Usage by Country")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```



##### Countries with higher CO2 emissions (e.g., those on the left) generally tend to have lower renewable energy percentages, suggesting a reliance on fossil fuels.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="Renewable_Energy_pct",
    y="CO2_Emissions_tons_per_capita",
    hue="Country"
)
plt.title("Scatter Plot: Renewable Energy % vs CO2 Emissions per Capita")
plt.xlabel("Renewable Energy (%)")
plt.ylabel("CO2 Emissions (tons per capita)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
